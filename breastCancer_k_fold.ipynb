# K-fold cross validation
is a statistical method used to estimate the skill of machine learning models. Here’s how it works:

- Data Splitting: You start with your dataset. Instead of just splitting it into a training set and a test set once, you divide it into k equally sized parts (or “folds”). For example, if k=5, you have 5 folds.
- Training and Testing: You train your model k times. In each iteration:
  - One fold is used as the test set.
  - The remaining k-1 folds are used as the training set.
  - You evaluate the model’s performance on the test set.
- Skill Estimation: By repeating this process k times (with different folds as the test set each time), you get a more reliable estimate of how well your model will perform on unseen data. It helps you avoid overfitting and gives you a better understanding of your model’s generalization ability.

```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
import tensorflow as tf
print("TensorFlow version:", tf.__version__)

import keras
print("Keras version:", keras.__version__)
```

```python
features_cancer = pd.read_csv('entradas_breast.csv')
classes_cancer = pd.read_csv('saidas_breast.csv')
```

```python
features_cancer.shape
```

```python
def classifier_model_dense():
  classifier = Sequential()
  classifier.add(Dense(units = 16, activation = 'relu',
                      kernel_initializer='random_uniform',
                      input_dim= 30))
  classifier.add(Dense(units = 16, activation = 'relu',
                      kernel_initializer='random_uniform'))
  classifier.add(Dense(units = 1, activation = 'sigmoid'))
  optimizer = keras.optimizers.Adam( learning_rate = 0.001,
                                    decay=0.001,
                                    clipvalue=0.5)
  classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy',
                    metrics = ['binary_accuracy'])

  return classifier
```

```python
#cv = is the number of K-fold, meaning it takes little time to train 10 times
# Notice we didn’t split the previous data into the train and test sets
classifier = KerasClassifier(build_fn=classifier_model_dense,
                             batch_size = 10,
                             epochs = 100)
result = cross_val_score(estimator = classifier,
                         X = features_cancer,
                         y = classes_cancer,
                         cv = 10,
                         scoring = 'accuracy')
```
